```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
from skimage import io, measure, filters, morphology, color, exposure, segmentation, util
import scipy.ndimage as nd
import numpy as np
```

```python
img1 = io.imread("data/crowd1.gif")
print img1.shape
```

```python
gray_frames1 = np.array([ filters.median(color.rgb2gray(f), selem=morphology.disk(1.)) for f in img1])
print gray_frames1.shape
```

```python
## try to get the background image by using a rank filter over different time frames?
bg_img = np.percentile(gray_frames1, q = 90, axis = 0, )
bg_img = exposure.equalize_hist(bg_img)
plt.imshow(bg_img, cmap = plt.cm.gray)
```

```python
labeled_frames = []
## play with parameters, pipes ...
## observations: 
## a. try to equalize the images because the light condition changes over time
## b. use the absolute value of difference between some people wear dark clothes
## c. try to add more steps in the pipe one by one, from simplest - interactive plotting
## d. morphology is important because we want to detect verticl-box as human shapes
for i in xrange(img1.shape[0]):
    cur_img = gray_frames1[i, ...]
    cur_img = exposure.equalize_hist(cur_img, )
    
    ## smooth out changes from background
    fg_img = np.abs(cur_img - bg_img)#filters.gaussian_filter(np.abs(cur_img - bg_img), 1.)
    
    ## pick the siginifcant changes from background
    thr = np.percentile(fg_img.ravel(), 85)#filters.threshold_otsu(fg_img)#
    biimg = fg_img > thr
    biimg = filters.median(biimg, selem = morphology.disk(2.)) ## remove smaller parts
    verticle_connector = np.zeros((15, 3)); verticle_connector[:, 1] = 1 ## vertically connect
    biimg = morphology.binary_closing(biimg, selem=verticle_connector)
    horizontal_separator = np.zeros((3, 3)); horizontal_separator[1, :] = 1 ## horizontally disconnect
    biimg = morphology.binary_opening(biimg, selem = horizontal_separator, )
    #biimg = nd.binary_fill_holes(biimg)
    
    
    #biimg = morphology.closing(biimg, selem=morphology.disk(2.))
    
    labels = measure.label(biimg)
    
    ## rought critera - could be improved
    regions = measure.regionprops(labels)
    selected_labels = [r.label for r in regions if (r.area > 100) 
                                                if (np.pi/4 <= r.orientation)
                                                #if r.eccentricity > 0.5
                                                #if (r.bbox[2]-r.bbox[0])*1. /(r.bbox[3]-r.bbox[1])> 1.
                      ]
    newlabels = np.zeros_like(labels)
    for ilabel, label in enumerate(selected_labels, 1):
        newlabels[labels==label] = ilabel
    
    labeled_image = color.label2rgb(newlabels, image=img1[i], alpha = 0.2)
    labeled_frames.append(labeled_image)
    #plt.figure(figsize = (3, 3))
    #plt.imshow(labeled_image)
    #plt.title(i)
    
```

```python
fig, axes = plt.subplots(13, 10, figsize = (1.5*10, 1.5*13))
fig.subplots_adjust(wspace=0, hspace=0, left=0,  bottom=0, )
axes = axes.ravel()
for frame, ax in zip(labeled_frames, axes):
    ax.set_axis_off()
    ax.imshow(frame)
```

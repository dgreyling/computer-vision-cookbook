## Estimate the river area
- [Original article](http://4quant.com/Flood-Risk/) 4quant.com has a good application of estimating flood effect via a time series modelling
- what we do here is a simple case study of how to detect the river area in a static image
- [original image data](http://4quant.com/Flood-Risk/)

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
from skimage import io, measure, filters, morphology, feature
import numpy as np
from scipy import ndimage as nd
```

```python
## read image
img = io.imread("data/river.png")
plt.imshow(img)
```

```python
## detect river as deep blue segements
blues = img[:, :, 2].ravel()
_ = plt.hist(blues, bins = 50)

blue_lower, blue_upper = 55, 65
binary_image = 1 - ( img[:, :, 2] > blue_lower) & (img[:, :, 2] < blue_upper)
binary_image = binary_image.astype(np.int)
binary_image = morphology.binary_opening(binary_image, selem = morphology.disk(2.5))
#binary_image = morphology.binary_closing(binary_image, selem = morphology.disk(2.5))

plt.figure()
plt.imshow(binary_image, cmap = plt.cm.gray)
```

```python
## segments
distance = nd.distance_transform_cdt(binary_image, )
peaks = feature.peak_local_max(distance, indices=False)
peak_labels = measure.label(peaks)
label_image = morphology.watershed(-distance, peak_labels, mask=binary_image)
plt.imshow(label_image, cmap=plt.cm.spectral)
regions = measure.regionprops(label_image)
```

```python
def is_river_region(region):
    r, c = region.centroid
    e = region.solidity
    return (90 < c <150) #and (region.area > 200)

selected_regions = [r for r in regions
                   if is_river_region(r)]

relabel_image = np.zeros_like(label_image)
for r in selected_regions:
    relabel_image[label_image==r.label] = 1
    
plt.imshow(relabel_image)
```

***Obviously the result is not good enough, try another way of doing segmenation**

```python
edges = feature.canny(binary_image, sigma = .5)
plt.imshow(edges)
label_image = morphology.watershed(edges, measure.label(edges), mask = binary_image)
plt.figure()
plt.imshow(label_image, cmap = plt.cm.spectral)
regions = measure.regionprops(label_image)
```

```python
def is_river_region(region):
    r, c = region.centroid
    e = region.solidity
    return (90 < c <150) and (region.area > 900)

selected_regions = [r for r in regions
                   if is_river_region(r)]

relabel_image = np.zeros_like(label_image)
for r in selected_regions:
    relabel_image[label_image==r.label] = 1
    
plt.imshow(relabel_image)

## fill the holes
filled_labels = nd.binary_fill_holes(relabel_image)
plt.figure()
plt.imshow(filled_labels)
print "area of river is", filled_labels.sum()
```

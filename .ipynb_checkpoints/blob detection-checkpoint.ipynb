### Blob detection
- [original discussion](http://scikit-image.org/docs/dev/auto_examples/plot_blob.html)
- blob detection is important when counting, detecting regions or esimating rough shape of the segments. In the article above, Blobs are defined as bright on dark or dark on bright regions in an image.
- skimage advocates three main approaches
    - Laplacian of Gaussian (LoG)
        - pros: accurate for very small blobs
        - cons: significantly slower for larger blobs; only detects bright blobs on dark
    - Difference of Gaussian (DoG)
        - pros: accurate for very small blobs
        - cons: slower for larger blobs; only detects bright blobs on dark
    - Determinant of Hessian (DoH)
        - pros: detection speed indepedent of blob size; detect both bright on dark and dark on bright
        - cons: small blobs (<3px) are not detected accurate
    - all these methods work with same interface: returns (r, c, value)
- [good explaination of how blob detections work](http://www.cs.utah.edu/~jfishbau/advimproc/project1/)
        
        
we use another image different from the original one to show the technique

```python
%matplotlib inline
```

```python
from skimage import io, color, exposure, feature, draw
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import patches
```

```python
## load the image
snow = io.imread("http://www.bogotobogo.com/Matlab/images/MATLAB_DEMO_IMAGES/snowflakes.png", as_grey=True)
plt.imshow(snow, cmap = plt.cm.gray)
```

```python
## enhance the contrast
enhanced = exposure.equalize_adapthist(snow, )
plt.imshow(enhanced, cmap = plt.cm.gray)
```

```python
log_rcv = feature.blob_log(enhanced, min_sigma=1, max_sigma=10, num_sigma=10, threshold=.1)
fig, ax = plt.subplots(1, 1, figsize = (12, 12))
ax.imshow(enhanced, cmap = plt.cm.gray)
ax.autoscale(False)
for r, c, v in log_rcv:
    ax.add_patch(patches.Circle((c, r), v*np.sqrt(2), fill=False, color = "red"))
```

```python
dog_rcv = feature.blob_dog(enhanced, min_sigma=1, max_sigma=10,  threshold=.1, )
fig, ax = plt.subplots(1, 1, figsize = (12, 12))
ax.imshow(enhanced, cmap = plt.cm.gray)
ax.autoscale(False)
for r, c, v in dog_rcv:
    ax.add_patch(patches.Circle((c, r), v*np.sqrt(2), fill=False, color = "red"))
```

```python
doh_rcv = feature.blob_doh(enhanced, min_sigma=1, max_sigma=10, num_sigma=10, threshold=.01, )
fig, ax = plt.subplots(1, 1, figsize = (12, 12))
ax.imshow(enhanced, cmap = plt.cm.gray)
ax.autoscale(False)
for r, c, v in doh_rcv:
    ax.add_patch(patches.Circle((c, r), v*np.sqrt(2), fill=False, color = "red"))
```

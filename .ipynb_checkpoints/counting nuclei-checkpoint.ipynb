# Counting Nuclei
It is one of the "hello world" examples in computer vision / image processing tasks, which has been discussed in different tutorials. One of them was from the authors of mahotas. The article is online [here](http://pythonvision.org/basic-tutorial/)
- It illustrates very basic steps in a lot of image processing tasks
    - filtering to remove noise
    - thresholding based on properties of ROI (e.g., color, intensity and etc)
    - morphology (opening closing to make the regions in better shapes)
    - labelling and filtering of labels based on area/oritentation and etc
    - present the results
- the above steps should be iterative
- it illustrates the nature of most image processing tasks - image dependent and the implementation is almost based on human inspection, and then by some algorithmic code
- the data is from [here](http://pythonvision.org/media/files/images/dna.jpeg). The original code is using mahotas, I am using skimage here

```python
%matplotlib inline
```

```python
import numpy as np
import matplotlib.pyplot as plt
import mahotas as mh
from skimage import io, filters, morphology, color, measure, feature
```

```python
## load data
dna = io.imread("http://pythonvision.org/media/files/images/dna.jpeg")
plt.imshow(dna, cmap = plt.cm.gray)
```

## first attempt

```python
## noise filtering - not really necessary here?
## thresholding - a first thing to try if you got no idea is otsu
bi_dna = dna > filters.threshold_otsu(dna)
plt.imshow(bi_dna, cmap = plt.cm.gray)
```

```python
## now you see one of the region is too scattered
## usually you can use a gaussian filter to make it blur first
## or use morphlogy to close certain regions

## try the closing first
## but you can see it comes with a price, because when we close the gaps within an object
## at the same time it will close the boundary between different objects.
## so we have two different needs here - (a) within object, make it closed, (b) inbetween objects, make it open
## so we can try gaussian bluring at this moment - but in practice it is usually a judgmental call

blur_dna = morphology.binary_closing(bi_dna, selem=morphology.disk(50))
plt.imshow(blur_dna, cmap= plt.cm.gray)
```

```python
## thresholding the with gaussian filter
blur_dna = filters.gaussian_filter(dna, 15)
bi_dna = blur_dna >= filters.threshold_otsu(blur_dna)
plt.imshow(bi_dna, cmap = plt.cm.gray)
```

```python
## make them seperate by opening
open_dna = morphology.binary_opening(bi_dna, selem=morphology.disk(25))
plt.imshow(open_dna, cmap=plt.cm.gray)
labels = measure.label(open_dna)
plt.figure()
plt.imshow(labels, cmap = plt.cm.spectral)
```

```python
## 0 is background in labels
print "we found %i nuclei" % (labels.max())
```

## watershed method
- obviously the previous code didnt achieve the purpose - counting the nuclei that seem to be glued together
- the morphology.openining has a limit for these tasks
- we need better segmentation methods

```python
## to see why it is hard, take a look at the contours - they are glued together
## and finding contour or convex hull is usually the last step to presentation
contours = measure.find_contours(bi_dna, level = 0.5, )
plt.imshow(dna)
plt.autoscale(False)
for contour in contours:
    plt.plot(contour[:, 1], contour[:, 0], "r-")
```

```python
## so the main challenge here is that certain regions are glued 
## and their boundaries/edges based on intensity are not very clear, 
## so thresholding based segmentation may fail on this case

## HOWEVER, notice that the regions are still distinguishable if we view them as
## regions like hills (with one of several peaks). We can count them by watershed
## if we manage to fit one peak in each region
## WATERSHED needs distance map and local maxima as input!

# WATERSHED steps
# find the distance map - distance of image pixels to background (value 0)
import scipy.ndimage as nd
distance = nd.distance_transform_edt(bi_dna, )
plt.figure()
plt.imshow(distance)
plt.title("distance map to background")
# find local max as center of nuclei 
# DO INCLUDE BORDER for the half-shown nuclei on the bottom border
peaks = feature.peak_local_max(distance, indices=False, exclude_border=False)
peak_rs, peak_cs = np.where(peaks > 0)
plt.figure()
plt.imshow(bi_dna)
plt.autoscale(False)
plt.plot(peak_cs, peak_rs, "g*")
plt.title("local peaks found")
# label the peaks as basin for watershed
peak_labels = measure.label(peaks)
# watershed - high value will be barrier, putting markers on low value region to make sure it is found
labels = morphology.watershed(-distance, markers = peak_labels, mask=bi_dna, )
plt.figure()
plt.imshow(labels)
## if you want to generate the Voronoi by not using a mask image
plt.figure()
plt.imshow(morphology.watershed(-distance, markers=peak_labels, ))
```

```python
## the above result is not expected, because there are artificial borders
## beside those real ones between nucleis. Gaussian filter should blur this out / or closing/opening
## REDO WITH BLUR IMAGE

import scipy.ndimage as nd
selem = morphology.disk(11)
blur_dna = morphology.closing(morphology.opening(dna, morphology.disk(7)), selem)
#blur_dna = filters.median(dna, morphology.disk(11))
bi_dna = blur_dna > filters.threshold_otsu(blur_dna)
distance = nd.distance_transform_edt(bi_dna, )
plt.figure()
plt.imshow(distance)
plt.title("distance map to background")
# find local max as center of nuclei 
# DO INCLUDE BORDER for the half-shown nuclei on the bottom border
peaks = feature.peak_local_max(distance, indices=False, exclude_border=False)
peak_rs, peak_cs = np.where(peaks > 0)
plt.figure()
plt.imshow(bi_dna)
plt.autoscale(False)
plt.plot(peak_cs, peak_rs, "g*")
plt.title("local peaks found")
# label the peaks as basin for watershed
peak_labels = measure.label(peaks)
# watershed - high value will be barrier, putting markers on low value region to make sure it is found
labels = morphology.watershed(-distance, markers = peak_labels, mask=bi_dna, )
plt.figure()
plt.imshow(labels)
## if you want to generate the Voronoi by not using a mask image
plt.figure()
plt.imshow(morphology.watershed(-distance, markers=peak_labels, ))
```

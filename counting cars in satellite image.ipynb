- Algorithm description
    - segmentation in images by finding useful properties of cars: the demo [Counting Cars Demo from 4quant](http://4quant.com/countingcarsdemo/) uses the average intensity as the feature here. $intensity = \frac{red+blue+green}{3}$, since cars are reflective and have many different colors
    - labelling
    - filtering by area and shape
    - counting

```python
%matplotlib inline
```

```python
from skimage import io, morphology, color, measure, filters
import matplotlib.pyplot as plt
import numpy as np
```

```python
## load image
image1 = io.imread("http://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/18/91777/137204")
plt.figure(figsize = (10, 10))
plt.imshow(image1)
```

```python
## preprocessing - denoising
filtered_image = filters.gaussian_filter(image1.mean(axis = 2), sigma = .5)
#filtered_image = morphology.closing(filtered_image, selem = morphology.disk(2))

## binarize image| by using intensity
intensity_thr = 140 # trial and error
_ = plt.hist(image1.mean(axis=2).ravel(), bins = 25)
plt.title("intensity_thr looks good to distinguish shinny things")
binary_image1 = ( filtered_image > intensity_thr)
plt.figure(figsize = (10, 10))
plt.imshow(binary_image1, cmap = plt.cm.gray)
```

```python
## label the image
#binary_image1 = morphology.binary_closing(binary_image1, selem = morphology.disk(2.))
image1_labels = measure.label(binary_image1)
plt.figure(figsize = (10, 10))
plt.imshow(image1_labels, cmap = plt.cm.spectral)
print image1_labels.max()
```

```python
## filtering the labels based on two conditions
## 1. region area between (130, 220)
## 2. region circularity between (.25, .4)
regions = measure.regionprops(image1_labels)
selected_regions = [r for r in regions if 10<r.area<70 and r.eccentricity>.8 and r.solidity > 0.7]
detected_image1 = np.zeros_like(image1_labels)
for i, r in enumerate(selected_regions, 1):
    detected_image1[image1_labels==r.label] = i
plt.imshow(detected_image1, cmap = plt.cm.spectral)
```

```python
## it misses some cars and is more confident with finding those SHINNY cars
plt.figure(figsize = (8, 8))
plt.imshow(image1)
plt.autoscale(False)
for r in selected_regions:
    y, x = r.centroid
    plt.plot(x, y, "r.")
```

## Doing OCR with skimage and tesseract 

- [original discussion](http://stackoverflow.com/questions/31436514/good-image-but-no-text-coming-from-ocr-why-python-skimage-pil-tesseract)
- approach
    - process the images to make it clear - e.g. binarize it
    - resize the roi (according to the discussion, this is crucial for tesseract to be successuful)
    - get the OCR

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
from skimage import io, color, filters, measure, morphology, feature, util, transform
import scipy.ndimage as nd
import numpy as np
import pytesseract
from PIL import Image
```

```python
## get the image
img = io.imread("http://i.stack.imgur.com/WfhA6.jpg")
plt.imshow(img)
```

```python
## process the data
gray_img = color.rgb2gray(img)
binary = gray_img > filters.threshold_otsu(gray_img)
plt.imshow(binary, cmap = plt.cm.gray)
```

```python
## find the ROI by cropping
roi = binary[1000:2000, 1800:2800]
plt.imshow(roi)
```

```python
## based on the original discussion, sizing is the key to OCR with tesseract

## pytesseract (like many other external image tools) assumes images to be in PIL.Image format
## so convertion is in place
def array_to_img(arr):
    return Image.fromarray(util.img_as_ubyte(arr))


resized_roi = transform.resize(roi, (200, 200))
plt.imshow(resized_roi)
print pytesseract.image_to_string(array_to_img(resized_roi))
```

```python
## try different sizes - it depends on the image
for scale in np.arange(0.05, 1.5, 0.05, ):
    resized_roi = transform.rescale(roi, scale, )
    print "=" * 10
    print scale, resized_roi.shape
    print pytesseract.image_to_string(array_to_img(resized_roi))
```

## looks like the optimal sizes range from 200 to 500

## debluring an image 
- [skimage example](http://scikit-image.org/docs/dev/auto_examples/plot_restoration.html)
- [matlab example](http://www.mathworks.com/help/images/examples/deblurring-images-using-a-wiener-filter.html)

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
from skimage import io, restoration, util, transform, data, filters
import numpy as np
```

```python
img = data.camera()
plt.imshow(img, cmap = plt.cm.gray)
```

```python
## blurred with gaussian filter with 5
blur = filters.gaussian_filter(img, 7)
plt.imshow(blur, cmap = plt.cm.gray)
```

```python
## deblur with unsupervised wiener with smaller psf
fig, axes = plt.subplots(1, 3, figsize = (15, 5))
psf_sizes = [3, 4, 10]

for psf_size, ax in zip(psf_sizes, axes):
    psf = np.ones((psf_size, psf_size)) / psf_size
    deblur, _ = restoration.unsupervised_wiener(blur, psf, )
    ax.imshow(deblur, cmap = plt.cm.gray)
```
